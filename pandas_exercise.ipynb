```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
student_data = {
    "Class": ["A", "A", "B", "B", "A", "C", "C", "C", "A"],
    "Name": ["Prosperity", "Annie", "Martha", "Lisa", "Bimbo", "Jessica", "Mary-Ann", "James", "Kelvin"],
    "English_Language": [None,56, 80, 90, 44, 77, 33, 20, 89],
    "Mathematics": [23, 44, 73, 11, 76, 39, 22, 79, 40],
    "Basic_Science": [67, 88, 39, 54, 66, 39, 88, 80, None]
}
```

```python
df = pd.DataFrame(student_data)
df
```

```python
#for small number of columns
df["English_Language"] = df["English_Language"].fillna(df["English_Language"].mean())
df
```

```python
numeric_cols = df.select_dtypes(include="number").columns
df[numeric_cols]=df[numeric_cols].fillna(df[numeric_cols].mean())
df[numeric_cols] = df[numeric_cols].round(0)
df
```

### This analysis explores student scores to determine average performance, rankings, pass rates, and subject difficulty.

```python
df["Average"]= df[["English_Language", "Mathematics", "Basic_Science"]].mean(axis = 1).round(2)
df
```

#### Student names and average scores only

```python
df[["Name", "Average"]]
```

#### Top Students by Average

```python
# df_top_students = df[df["Average"]> 60][["Name", "Average"]]
# df_top_students = df_top_students.sort_values(by="Average", ascending = False)
# df_top_students
sorted_average = df[df["Average"] > 60][["Name", "Average"]].sort_values(by="Average", ascending =False)
sorted_average
```

```python
df.sort_values(by= "Average", ascending=True, inplace = True)
df
```

```python
class_group = df.loc[df.groupby("Class")["Average"].idxmax()]
class_group
```
